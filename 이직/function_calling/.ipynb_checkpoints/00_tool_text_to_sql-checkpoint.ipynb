````python
# ==========================================================
# dbpool 전역 처리
# 20260209 dbpool 전역 처리 + 비동기 처리
# 류성원
# ==========================================================

import logging
import asyncio
from collections import defaultdict
from typing import Dict, List, Set, Optional, Tuple

import httpx
import oracledb
import pandas as pd

# Oracle 접속 정보
ORACLE_CONFIG = {
    "user": "intraware_999",
    "password": "intraware_lnk",
    "dsn": (
        "(DESCRIPTION="
        "(FAIL_OVER=ON)(LOAD_BALANCE=OFF)"
        "(ADDRESS=(PROTOCOL=TCP)(HOST=172.16.2.63)(PORT=1526))"
        "(ADDRESS=(PROTOCOL=TCP)(HOST=172.16.2.65)(PORT=1526))"
        "(CONNECT_DATA=(SERVICE_NAME=EISDB))"
        "(FAIL_OVER_MODE=(TYPE=SELECT)(METHOD=BASIC)))"
    ),
    "min_connections": 2,
    "max_connections": 10,
    "increment": 1,
}

# ----------------------------------------------------------------------------------------------------------
# 전역 커넥션 풀
# ----------------------------------------------------------------------------------------------------------

_pool: Optional[oracledb.ConnectionPool] = None
_pool_lock = asyncio.Lock()

async def get_oracle_pool() -> oracledb.ConnectionPool:
    global _pool

    if _pool is not None:
        return _pool

    def _create_pool():
        return oracledb.create_pool(
            user=ORACLE_CONFIG["user"],
            password=ORACLE_CONFIG["password"],
            dsn=ORACLE_CONFIG["dsn"],
            min=ORACLE_CONFIG["min_connections"],
            max=ORACLE_CONFIG["max_connections"],
            increment=ORACLE_CONFIG["increment"],
        )
    try:
        _pool = await asyncio.to_thread(_create_pool)
        print(
            f"Oracle connection pool created: "
        )
    except Exception as e:
        print(f"Oracle pool creation failed: {e}", exc_info=True)
        _pool = None
        raise

    return _pool

async def close_oracle_pool():
    """앱 종료 시 풀 정리"""
    global _pool
    if _pool is not None:
        await asyncio.to_thread(_pool.close, force=True)
        _pool = None
        print("Oracle connection pool closed")

async def execute_query(sql: str, params: dict = None) -> pd.DataFrame:
    """
    커넥션 풀에서 커넥션을 빌려 쿼리 실행 후 반환
    Args:
        sql: 실행할 SQL
        params: 바인드 변수 (선택)
    Returns:
        결과 DataFrame
    """

    pool = await get_oracle_pool()
    if pool is None:
        raise ConnectionError("Oracle connection pool is not available")

    def _run_query():
        connection = None 
        try:
            connection = pool.acquire()
            cursor = connection.cursor()
            try:
                if params:
                    cursor.execute(sql, params)
                else:
                    cursor.execute(sql)

                column_names = [desc[0] for desc in cursor.description]
                rows = cursor.fetchall()
                return pd.DataFrame(rows, columns=column_names)
            except Exception as e:
                print(f"Query execution failed: {e}")
                raise
            finally:
                cursor.close()
        except Exception as e:
            print(f"DB operation failed: {e}")
            raise
        finally:
            if connection is not None:
                pool.release(connection)

    return await asyncio.to_thread(_run_query)
````

````python
async def get_response(messages: dict):
    payload = {
        "model":model_name,
        "messages": messages,
        "temperature": 0.9,
        "extra_body": {
            "repetition_penalty":1.1,
            "skip_special_tokens": True,
            "frequency_penalty": 0.1,
            "top_p": 0.95,
            "max_tokens": 4096,
            },
        "stream": True,
        "reasoning_effort": "medium",
        "tools": tools,
        "tool_choice": "required",
        }
    
    response = await client.chat.completions.create(**payload)

    out = display(Markdown(""), display_id=True)
    
    final_answer = []
    accumulated = ""
    reasoning_done = False
    content_started = False
    async for chunk in response:
        delta = chunk.choices[0].delta
        if hasattr(delta, "reasoning_content") and delta.reasoning_content:
            content = delta.reasoning_content
            final_answer.append(content)
            accumulated += content
            update_display(Markdown(accumulated), display_id=out.display_id)
            # print(content, end="", flush=True)
        
        if hasattr(delta, "content") and delta.content is not None:
            chunk_content = chunk.choices[0].delta.content

            if isinstance(chunk_content, str):
                if not reasoning_done and not content_started:
                    reasoning_done = True
                    content_started = True
                    # chunk_content = '<br><br>' + chunk_content

                chunk_content = chunk_content.replace('```html', '').replace('```', '')
                    
                final_answer.append(chunk_content)
                accumulated += chunk_content
                update_display(Markdown(accumulated), display_id=out.display_id)
                # print(chunk_content, end="", flush=True)

    return accumulated
````

````python
async def handle_draft_document(args: dict) -> str:

    system_content =  f"""
# 당신은 제공된 정보를 기반으로 문서 초안을 생성하는 AI 어시스턴트입니다.
아래 내용을 따라 최선을 다해 초안을 생성하여, 사용자에게 최고의 만족도를 줄 수 있도록 최대한 노력합니다.

### [기본 지침]
- **기본 양식이나 예시가 있는 경우:**
    - **제공된 양식을 유지하려고 최선을 다합니다.(가장 중요)**
    - **목차, 개요, 말투, 표(Table) 구조 등을 초안에 반영**합니다.
        - 페이지(Page) 정보 제외
        - 오타 및 띄어쓰기 제외하며, 맞춤법을 교정해서 초안에 반영합니다.
    - **절대로 양식에서 표가 아닌 부분을 표(Table)로 변환하지 않습니다.(원본 형식 유지 필요)**
- **문서 예시가 없는 경우:**
    - 유동적으로 작성하여 서론, 본문, 결론을 기본으로 하여 필요에 따라 섹션을 구분합니다.
    - 먼저 사용자의 요청에 따른 초안을 우선 제시하고, 이후 대화를 통해 필요한 정보를 수집합니다. 
- 명시적으로 언급되지 않은 한 기본값을 추정하지 마십시오.
- 핵심 내용에 집중하고 중복 표현이나 불필요한 설명은 배제합니다.
- **누락되었거나 불명확한 사항이 있을 경우 마무리 멘트 단계에서 사용자에게 명확히 물어보세요.**
- 별도의 요청이 없다면 공공기관 양식에 맞추어 명확하고 **공식적인 어조**로 작성되어야 합니다.
    - 이모티콘 사용 금지(예: 1️⃣, 2️⃣ 등)

### [주의 사항] - 강력한 패널티 적용 대상
- 사용자가 개인정보(예: 이름, 주소, 연락처)를 명시적으로 포함시키라는 지시가 없는 경우:
    - 해당 정보를 자동으로 마스킹(OOO 처리)하거나 삭제합니다.
- HTML Viewer 렌더링 생성(태그 간 줄바꿈 금지)
    - **절대 코드기호(```html, ``` 등)는 사용하지 않습니다.(사용 시 강력한 패널티 적용)**
    - 사용자는 개발자가 아닌 HTML 문법을 모르는 일반 사용자입니다.
        - HTML 태그에 관련된 내용은 절대 언급하지 않습니다. (예: <strong>, <em>, <div>, <style> 등)
        - 본문 내용에 대해서만 맞춤 수정으로 안내합니다.
- **초안 생성 시 문서 페이지 정보는 생성에서 제외**합니다.

#### [작업 순서]
1. 샘플 문서 양식 판단 여부
    - 샘플 문서 양식 사용 여부 판단 필수
2. 지침 사항에 따라 초안 생성
3. 마무리 멘트 안내(품질 개선을 위한 내용)(필수)
    - 메일 초안이 끝난 뒤 빈 줄을 두고 아래와 같이 동적으로 안내 문구를 삽입합니다.
    - 품질 개선을 위한 추가적인 제안을 안내합니다.
    - 불명확한 사항이 있을 경우 사용자에게 재질문합니다.

#### 샘플 문서 양식: 
{sample_document}

"""
    
    messages=[
        {"role": "system", "content": system_content},
        {"role": "user", "content": f"""{args['user_query']}
        """},
    ]
    return await get_response(messages)
````
