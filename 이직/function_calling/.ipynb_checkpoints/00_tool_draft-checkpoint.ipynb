````python
import asyncio
from openai import OpenAI, AsyncOpenAI
import json
from IPython.display import display, Markdown, update_display
from pymilvus import MilvusClient, AnnSearchRequest, RRFRanker, WeightedRanker
import collections

client = AsyncOpenAI(base_url='http://vllm-gpt-oss-120b-service.kamco.svc.cluster.local:8000/v1',api_key='1234')
model_name = "gpt-oss-120b"

embedding_model = 'pixie'
embedding_client = AsyncOpenAI(
    base_url='http://vllm-telepix-pixie-spell-preview-service.kamco.svc.cluster.local:8000/v1',
    api_key="EMPTY"
)

milvus_client = MilvusClient(
    uri="http://milvus.chatsam-milvus.svc.cluster.local:19530"
)
````

````python
async def get_response(messages: dict):
    payload = {
        "model":model_name,
        "messages": messages,
        "temperature": 0.9,
        "extra_body": {
            "repetition_penalty":1.1,
            "skip_special_tokens": True,
            "frequency_penalty": 0.1,
            "top_p": 0.95,
            "max_tokens": 4096,
            },
        "stream": True,
        "reasoning_effort": "medium"
        }
    
    response = await client.chat.completions.create(**payload)

    out = display(Markdown(""), display_id=True)
    
    final_answer = []
    accumulated = ""
    reasoning_done = False
    content_started = False
    async for chunk in response:
        delta = chunk.choices[0].delta
        if hasattr(delta, "reasoning_content") and delta.reasoning_content:
            content = delta.reasoning_content
            final_answer.append(content)
            accumulated += content
            update_display(Markdown(accumulated), display_id=out.display_id)
            # print(content, end="", flush=True)
        
        if hasattr(delta, "content") and delta.content is not None:
            chunk_content = chunk.choices[0].delta.content

            if isinstance(chunk_content, str):
                if not reasoning_done and not content_started:
                    reasoning_done = True
                    content_started = True
                    # chunk_content = '<br><br>' + chunk_content

                chunk_content = chunk_content.replace('```html', '').replace('```', '')
                    
                final_answer.append(chunk_content)
                accumulated += chunk_content
                update_display(Markdown(accumulated), display_id=out.display_id)
                # print(chunk_content, end="", flush=True)

    return accumulated
````

````python
async def handle_draft_document(args: dict) -> str:

    # # 그룹웨어 문서 양식 호출
    # apprform = await get_apprform()
    
    # # 관련 문서 양식 추출
    # reranked_docs = get_rerank(apprform)
    
    # # 양식 필터링
    # if reranked_docs:
    #     form_name_list = list(set(pd.DataFrame(reranked_docs).drop_duplicates("doc")['doc'].tolist()))
    #     form_name_list = '[' + ', '.join(f"'{escape_milvus_string(item)}'" for item in form_name_list) + ']'
    # else:
    #     form_name_list = ['통합기안양식', '통합기안양식(내부용)', '통합기안양식(새출발기금)', '통합기안양식(장기소액연체자지원재단)', '통합기안양식(기업지원금융)']
        
    query_dense = await embedding_client.embeddings.create(model=embedding_model, input=args['user_query'])
    dense_query = [d.embedding for d in query_dense.data][0]

    collection_name = 'approval_dense_db_pixie_0111'
    result = milvus_client.search(
        data=[dense_query],
        anns_field="dense_vector",
        collection_name=collection_name,
        output_fields=["orig_text", "title", "formname"],
        # filter=f"formname IN {form_name_list} AND start_page == 1 AND NOT title LIKE '%테스트%' AND NOT orig_text LIKE '%테스트%'",
        filter=f"start_page == 1 AND NOT title LIKE '%테스트%' AND NOT orig_text LIKE '%테스트%'",
        timeout=30,
    )
    document = [s for s in result[0] if s['distance'] >= 0.5]
    if result:
        sample_document = '\n\n---\n\n'.join([f"<document={idx} form='{chunk['entity']['orig_text']}'>" for idx, chunk in enumerate(document[:3], start=1)])
    else:
        sample_document = "없음"
    # print('추출된 문서:')
    # display(Markdown(sample_document))
    # print('\n\n')
    
    system_content =  f"""
# 당신은 제공된 정보를 기반으로 문서를 초안하는 AI 어시스턴트입니다.

## 지침 사항
### [기본 지침]
- 기본 양식이나 예시가 있는 경우:
    - **기본적으로 제공된 양식을 유지하려고 최선을 다합니다.(가장 중요)**
    - **사용자 질문과 관련된 양식인지 판단하여 사용 여부를 결정**합니다.
    - **목차, 개요, 말투, 표(Table) 구조 등을 초안에 반영**합니다.
- 문서 예시가 없는 경우:
    - 유동적으로 작성하여 서론, 본문, 결론을 기본으로 하여 필요에 따라 섹션을 구분합니다.
    - 먼저 사용자의 요청에 따른 초안을 우선 제시하고, 이후 대화를 통해 필요한 정보를 수집합니다. 
- 명시적으로 언급되지 않은 한 기본값을 추정하지 마십시오.
- 핵심 내용에 집중하고 중복 표현이나 불필요한 설명은 배제합니다.
- **누락되었거나 불명확한 사항이 있을 경우 항상 사용자에게 명확히 물어보세요.**
- 별도의 요청이 없다면 공공기관 양식에 맞추어 명확하고 **공식적인 어조**로 작성되어야 합니다.
    - 이모티콘 사용 금지(예: 1️⃣, 2️⃣ 등)

### [주의 사항] - 강력한 패널티 적용 대상
- 사용자가 개인정보(예: 이름, 주소, 연락처)를 명시적으로 포함시키라는 지시가 없는 경우:
    - 해당 정보를 자동으로 마스킹(OOO 처리)하거나 삭제합니다.
- HTML Viewer 렌더링 생성(태그 간 줄바꿈 금지)
    - **절대 코드기호(```html, ``` 등)는 사용하지 않습니다.(사용 시 강력한 패널티 적용)**
    - 사용자는 개발자가 아닌 HTML 문법을 모르는 일반 사용자입니다.
        - HTML 태그에 관련된 내용은 절대 언급하지 않습니다. (예: <strong>, <em>, <div>, <style> 등)
        - 본문 내용에 대해서만 맞춤 수정으로 안내합니다.
    
#### [작업 순서]
최종 답변에서 사용자에게 단계별 제공하며 구체적인 이유 제공
1. 샘플 문서 양식 판단 여부
    - 샘플 문서 양식 사용 여부 판단 필수
    - 사용자 질문과 샘플 문서 양식 부합 여부
2. 지침 사항에 따라 초안 생성
3. 마무리 멘트 안내(품질 개선을 위한 내용)(필수)
    - 메일 초안이 끝난 뒤 빈 줄을 두고 아래와 같이 동적으로 안내 문구를 삽입합니다.
    - 품질 개선을 위한 추가적인 제안을 안내합니다.
    - 불명확한 사항이 있을 경우 사용자에게 재질문합니다.

#### 샘플 문서 양식: 
{sample_document}


"""
    messages=[
        {"role": "system", "content": system_content},
        {"role": "user", "content": f"""{args['user_query']}
        주제: {args.get('topic', '스스로 판단해서 작성해주세요.')}
        내용: {args['user_query']}
        추가 요구사항: {args.get('requirements', '맞춤 수정을 위해 어떻게 할 수 있을지 알려주세요.')}
        """},
    ]
    return await get_response(messages)
````

````python
query = """2026년 ESG 역량강화 교육 참여 요청 초안 생성해줘"""
#포트 사용 승인 요청 초안 생성해줘
args = {}
args['user_query'] = f'''{query}'''

answer = await handle_draft_document(args)
````
