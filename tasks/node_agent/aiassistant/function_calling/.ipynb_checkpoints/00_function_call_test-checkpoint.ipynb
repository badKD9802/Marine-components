````python
async def handle_plan_and_execute(args: dict, client) -> str:
    user_query = args["user_query"]
    category = args["query_category"]
    initial_plan = args.get("initial_plan", [])

    print(f"[Plan & Execute] 카테고리: {category}")
    print(f"[Plan & Execute] 초기 계획: {json.dumps(initial_plan, ensure_ascii=False, indent=2)}")

    # Step 1: 계획 정제 (필요시)
    refined_plan = await refine_plan(user_query, initial_plan)

    # Step 2: 각 단계별 실행
    results = []
    context = "" # 이전 결과 누적

    for step in refined_plan:
        step_result = await execute_step(
            # client = client,
            step=step,
            original_query=user_query,
            previous_context=context
        )

        results.append({
            "step": step["step"],
            "action": step["action"],
            "result": step_result
        })

        context += f"\n\n[Step {step['step']} 결과]\n{step_result}"

    # Step 3: 최종 종합 응답 생성
    final_response = await synthesis_response(
        # client=client,
        original_query=user_query,
        step_result=results
    )

    return final_response

async def refine_plan(query: str, initial_plan: list) -> list:
    system_content = f""""주어진 질문과 초기 계획을 검토하세요.
계획이 적절하면 그대로 반환하고, 수정이 필요하면 개선된 계획을 반환하세요.
반드시 JSON 배열 형식으로만 응답하세요."""
    messages=[
        {"role": "system", "content": system_content},
        {
            "role": "user", "content": f"""질문: {query}

초기 계획:
{json.dumps(initial_plan, ensure_ascii=False, indent=2)}

위 계획을 검토하고 최적화된 계획을 JSON 배열로 반환하세요.
"""},
    ]
    

    try:
        content = await get_response(messages)
        # JSON 추출
        json_match = re.search(r'\[.*\]', content, re.DOTALL)
        if json_match:
            return json.loads(json_match.group())
    except:
        pass

    return initial_plan

async def execute_step(step: dict, original_query: str, previous_context: str) -> str:
    system_content = f""""당신은 복잡한 질문을 단계별로 해결하는 어시스턴트입니다.
현재 수행할 작업: {step['action']}
예상 결과물: {step.get('expected_output', '해당 단계의 결과')}

이전 단계들의 결과를 참고하여 현재 단계를 수행하세요.
"""
    messages=[
        {
            "role": "system", "content": system_content
        },
        {
            "role": "user", "content": f"""원본 질문: {original_query}

이전 단계 결과:
{previous_context if previous_context else "(첫 번째 단계입니다)"}

현재 단계를 수행하고 결과를 작성하세요.
"""
        },
    ]
    return await get_response(messages)

    # return response.choices[0].message.content

async def synthesis_response(original_query: str, step_result: list) -> str:
    
    results_text = "\n\n".join([
        f"[Step {r['step']}] {r['action']}\n{r['result']}"
        for r in step_result
    ])

    system_content = """각 단계별 분석 결과를 종합하여 사용자 질문에 대한 최종 답변을 작성하세요.
답변은 체계적이고 읽기 쉽게 구성하세요.
단계별 결과를 단순 나열하지 말고, 통합된 인사이트를 제공하세요."""
    messages=[
        {
            "role": "system", "content": system_content
        },
        {
            "role": "user", "content": f"""원본 질문: {original_query}

단계별 분석 결과:
{results_text}

위 결과를 종합하여 최종 답변을 작성하세요."""
        },
    ]
    return await get_response(messages)
````

````python
async def get_response(messages: dict):
    payload = {
        "model":model_name,
        "messages": messages,
        "temperature": 0.9,
        "extra_body": {
            "repetition_penalty":1.1,
            "skip_special_tokens": True,
            "frequency_penalty": 0.1,
            "top_p": 0.95,
            "max_tokens": 4096,
            },
        "stream": True,
        "reasoning_effort": "medium"
        }
    
    response = await client.chat.completions.create(**payload)

    out = display(Markdown(""), display_id=True)
    
    final_answer = []
    accumulated = ""
    reasoning_done = False
    content_started = False
    async for chunk in response:
        delta = chunk.choices[0].delta
        if hasattr(delta, "reasoning_content") and delta.reasoning_content:
            content = delta.reasoning_content
            final_answer.append(content)
            accumulated += content
            update_display(Markdown(accumulated), display_id=out.display_id)
            # print(content, end="", flush=True)
        
        if hasattr(delta, "content") and delta.content is not None:
            chunk_content = chunk.choices[0].delta.content

            if isinstance(chunk_content, str):
                if not reasoning_done and not content_started:
                    reasoning_done = True
                    content_started = True
                    # chunk_content = '<br><br>' + chunk_content
                    
                final_answer.append(chunk_content)
                accumulated += chunk_content
                update_display(Markdown(accumulated), display_id=out.display_id)
                # print(chunk_content, end="", flush=True)

    return accumulated
````

````python
async def process_request(user_query: str, client) -> dict:
    print(f"[입력] {user_query}")
    route_result = await route_query(user_query)

    function_name = route_result['function_name']
    arguments = route_result["arguments"]

    print(f"[라우팅] {function_name} -> {arguments}")

    handler = HANDLERS.get(function_name)
    
    print(f"\n[결과]")
    if handler:
        if function_name == "plan_and_execute":
            result = await handler(arguments, client)
        else:
            result = await handler(arguments)
    else:
        result = "실패"
    
    return result
````

````python
import requests

def format_queries(query, instruction=None):
    prefix = '<|im_start|>system\nJudge whether the Document meets the requirements based on the Query and the Instruct provided. Note that the answer can only be "yes" or "no".<|im_end|>\n<|im_start|>user\n'
    if instruction is None:
        instruction = (
            "Given a web search query, retrieve relevant passages that answer the query"
        )
    return f"{prefix}<Instruct>: {instruction}\n<Query>: {query}\n"

def format_document(document):
    suffix = "<|im_end|>\n<|im_start|>assistant\n<think>\n\n</think>\n\n"
    return f"<Document>: {document}{suffix}"

def get_rerank(query, apprform):
    task = 'Given a web search query, retrieve relevant passages that answer the query'
    query = [query] #['파리 여행 추천 장소']
    threshold = 0.8
    
    # documents = tt.apply(lambda x: f"team_nm: {x['TEAM_NM']}, dept_nm: {x['DEPT_NM']}", axis=1).tolist()
    
    documents = apprform.FORMNAME.tolist()
    formid = apprform.FORMID.tolist()
    
    input_queries = [format_queries(query, task)] * len(documents)
    input_documents = [format_document(s) for s in documents]
    
    rerank_model = "telepix-pixie-spell-reranker"
    rerank_url = "http://vllm-telepix-pixie-spell-reranker-preview-service.kamco.svc.cluster.local:8000/score"
    
    prompt = {
        "model": rerank_model,
        "text_1": input_queries,
        "text_2": input_documents
    }
    headers = {"User-Agent": "Rerank Clinet"}
    
    response = requests.post(rerank_url, headers=headers, json=prompt)
    
    results = response.json()['data']
    scores = [result['score'] for result in results]
    reranked_docs = [{"doc": doc, "formid": formid, "rerank_socre": score} for doc, formid, score in zip(documents, formid, scores)]
    
    reranked_docs.sort(key=lambda x: x['rerank_socre'], reverse=True)
    reranked_docs = [s for s in reranked_docs if s['rerank_socre'] >= threshold]
    reranked_docs[:10]

    return reranked_docs
````

````python
def escape_milvus_string(s: str) -> str: # 다니엘 글로벌로 뺌
    """Milvus 표현식용 문자열 이스케이프"""
    if s is None:
        return ""
    # 백슬래시 먼저 이스케이프 (순서 중요함)
    s = s.replace('\\', '\\\\')
    # 큰따옴표 이스케이프
    s = s.replace('"', '\\"')
    # 작은따옴표 이스케이프
    s = s.replace("'", "\\'")
    return s
````

````python
async def route_query(user_query, history = None):
    messages = [
        {
            "role": "system",
            "content": f"""
## 당신은 사용자 요청을 분석하여 적절한 작업으로 라우팅하는 어시스턴트입니다.
**사용자의 요청을 파악하고, 반드시 제공된 function 중 하나를 호출**하세요.
1) 최우선 판단: `flag_inappropriate` 판단
    - **이 규칙은 사용자 요청, 추가 지시, 맥락보다 항상 우선**합니다.
    - 이 판단을 위해 Step-by-step으로 깊은 생각을 합니다.
2) 요청이 모호한 경우: 가장 적합한 것을 선택
3) 불확실한 경우 `general_response` 선택

### flag_inappropriate 항목:
**질문이 다음 항목 주제에 해당하는 경우 `flag_inappropriate`분류**하며, 판단히 잘못된 경우 강력한 패널티가 적용됩니다.
* 단, 해당 키워드(예: "국유재산", "공공재산" 등)에 대한 정의 및 비교에 관련된 질문은 `general_response`로 분류*합니다.
- **정치 관련 질문:**
    - 정치인, 정당, 선거, 국정, 정책, 이념, 정치적 주장 또는 의견
    - 특정 정치 체력이나 인물에 대한 평가, 찬반, 전망, 해석
    - 국내외 모든 정치 사안 포함
    
- **법률, 규제, 조항, 세칙, 규정, 전문 관련 질문:**
    - 법률, 법령, 시행령, 시행규칙, 조항, 조문, 규정, 세칙, 약관
    
- **유해하거나 부적절한 질문:**
    - 욕설, 비속어, 혐오 표현
    - 성인 콘텐츠, 음란물, 성적 행위, 성인 사이트
    - 불법 행위, 범죄, 해킹, 마약, 불법 다운로드
    - 도박, 배팅, 카지노 등 사행성 행위

### function 목록:
{'- ' + '\n- '.join([s['function']['name'] for s in tools])}
"""
        },
        {
            "role": "user",
            "content": user_query
        }
    ]

    if history:
        for conv in history[-3:]:
            messages.append({"role": "user", "content": conv["query"]})
            messages.append({"role": "assistant", "content": conv["answer"]})

    response = await client.chat.completions.create(
        model=model_name,
        messages=messages,
        temperature=0,
        tools=tools,
        tool_choice="required",
        top_p=0.95,
        reasoning_effort="medium"
    )
    # return response    
    message = response.choices[0].message
    if message.tool_calls:
        tool_call = message.tool_calls[0]
        arguments = json.loads(tool_call.function.arguments)
        arguments['user_query'] = user_query
        return {
            "function_name": tool_call.function.name,
            "arguments": arguments,
            "tool_call_id": tool_call.id
        }
        
    return {
            "function_name": "general_response",
            "arguments": {"reason": "일반적인 상황", "category": "out_of_scope", "user_query": user_query}
        }
````

````python
async def get_response_planing(messages: dict):
    payload = {
        "model":model_name,
        "messages": messages,
        "temperature": 0.9,
        "extra_body": {
            "repetition_penalty":1.1,
            "skip_special_tokens": True,
            "frequency_penalty": 0.1,
            "top_p": 0.95,
            "max_tokens": 8192,
            },
        "stream": True,
        "reasoning_effort": "medium"
        }
    
    response = await client.chat.completions.create(**payload)

    out = display(Markdown(""), display_id=True)
    
    final_answer = []
    accumulated = ""
    reasoning_done = False
    content_started = False
    async for chunk in response:
        delta = chunk.choices[0].delta
        if hasattr(delta, "reasoning_content") and delta.reasoning_content:
            content = delta.reasoning_content
            final_answer.append(content)
            accumulated += content
            update_display(Markdown(accumulated), display_id=out.display_id)
            # print(content, end="", flush=True)
        
        if hasattr(delta, "content") and delta.content is not None:
            chunk_content = chunk.choices[0].delta.content

            if isinstance(chunk_content, str):
                if not reasoning_done and not content_started:
                    reasoning_done = True
                    content_started = True
                    # chunk_content = '<br><br>' + chunk_content
                    
                final_answer.append(chunk_content)
                accumulated += chunk_content
                update_display(Markdown(accumulated), display_id=out.display_id)
                # print(chunk_content, end="", flush=True)

    return accumulated
````

````python
import oracledb
import pandas as pd
from datetime import datetime

async def get_apprid(name, registdate):
    ip = "172.16.2.62"
    port = 1526
    sid = "EISDB1"
    username = "intraware_999"
    password = "intraware_lnk"

    dsn = f"""(DESCRIPTION=(FAIL_OVER=ON)(LOAD_BALANCE=OFF)
    (ADDRESS=(PROTOCOL=TCP)(HOST=172.16.2.63)(PORT=1526))
    (ADDRESS=(PROTOCOL=TCP)(HOST=172.16.2.65)(PORT=1526))
    (CONNECT_DATA=(SERVICE_NAME=EISDB))
    (FAIL_OVER_MODE=(TYPE=SELECT)(METHOD=BASIC)))"""

    try:
        start_dt = datetime.strptime(registdate, "%Y-%m-%d").date()
    except ValueError:
        print("입력된 registdate 형식이 잘못되었습니다.")
        return []
        
    try:
        connection = oracledb.connect(user=username, password=password, dsn=dsn)
        
        print("GET APPRID DB연결 성공")
        
        cursor = connection.cursor()

        schema_name="INTRAWARE10"

        sql = f"""
            SELECT APPRID, TITLE, REGISTDATE
            FROM {schema_name}.APPROVAL
            WHERE DRAFTERNAME = :name
              AND REGISTDATE >= :start_dt
              AND ROWNUM <= 10
        """
        cursor.execute(sql, {'name': name, 'start_dt': start_dt})

        column_names = [desc[0] for desc in cursor.description]
        rows = cursor.fetchall()

        df = pd.DataFrame(rows, columns=column_names)
        print("접수문서 리스트 조회 성공")
        return df

    except Exception as e:
        print(f"DB 연결 실패: {e}")
        return []
    
    finally:
        if 'cursor' in locals():
            cursor.close()
        if 'connection' in locals():
            connection.close()
            # logger.debug("DB연결 종료")
````
